```python
#Step 1: 3X3 board representation 
from IPython.display import clear_output

def display_board(board):
    
    print(board[1], board[2], board[3])
    print(board[4], board[5], board[6])
    print(board[7], board[8], board[9])
```

```python
# Step 2 : Player input adn marker assignment 
def player_input():
    
    position = 'False'

    within_range = False

    in_range = range(10) #int 


    while position.isdigit() == False or within_range == False:

        position = input('Enter a number between (0-10)') #string

        #Checks digit as well as range 
        if position.isdigit() == True and int(position) in in_range: 

            within_range= True
            return int(position) 
    
        else: 
            #Tells precisely what is wrong 
            if position.isdigit() == False:
                within_range = False 
                print("Not a Number! Enter again!")
            elif int(position) not in in_range:
                within_range = False
                print("Number Not in Range! Enter again!")
```

```python
#Step 3: place marker on desired position
def place_marker(board, marker, position):
    
    board[position] = marker
    display_board(board)
```

```python
#Step 4: Winning conditions
def win_check(board, mark): #list of triplets

    #print("In wincheck!") 
    #print("Mark:", mark)
    v = board[1],board[2],board[3]==mark ; m = board[2]==board[5]==board[8]==mark    ; o = board[3]==board[6]==board[9]==mark
    
    y = board[1]==board[4]==board[7]==mark ; n = board[4]==board[5]==board[6]==mark    ; p = board[3]==board[5]==board[7]==mark

    z = board[1]==board[5]==board[9]==mark ; k =  board[7]==board[8]==board[9]==mark    

    winlist = [v, y, z, m, n, k, o, p] 
    #print(winlist)
    countx=0
    counto=0
    
    for  wincheck in winlist:
        if wincheck ==True:
            countx+=1

        elif wincheck == False:
            counto+=1

        #else :
            #print("Both lost! Play again!")
        
    if countx>=1: 
        #print(f"{mark} is a winner!")
        return True
    else: return False
```

```python
# Step 6: Check space availability on board
def space_check(board, position):

    empty_space=[]

    for index, item in enumerate(board, start=0):
        if item == ' ':
            empty_space.append(index) #stores list element in list 
            #print("item at index: ",index)
        
    if position in empty_space:
        #print("position is empty!")
        return True
    else:
        #print("Position is not empty!")
        return False 

    #print(f"Empty list: {empty_space}")
    
#space_check(test_board, 1)
```

```python
# Step 7: check whether board is full
def full_board_check(board):
    #call function 9 times to check weather 
    i=0; position=0; count=0; countempty=0
    #print(board)
    
    while i<10: #and full ==False:#need flag Calles space_check function 10 times and update position 
        
        x= space_check(board, position )
        position+=1
        
        #print("check x: ", x )
        if x==True:
            countempty+=1 

           # print(f"{position} is empty!")
            #return False
        else:
            count+=1 
            #print(f"{position} is full!") 

        i+=1
        
    #print(f"{countempty} spaces are empty!")
    #print(f"{count} spaces are full!")

    if count >=8:#min prob of winning a game 
        #print(f"\ncount: {count} Board is full!")
        #clear_output(wait=True)
        return True
        
#test_board = [' X','X',' X','O','O','X ','O ',' ',' ','X '] 
#full_board_check(test_board)
```

```python
# Step 8: Player's next input based on free position available on board 
def player_choice(match_board):
    #print("Enter you choice")
    
    Free = False
    
    while Free == False:
        choice = player_input()
        Free = space_check(match_board, choice)
        if Free == True:
            print("Position is Free! Choice is saved!")
            return choice 
        else:
            print("Position is full! Enter again!")
```

```python
# Step 9: replay 
def replay():
    
    ans= input("Do you wish to play again? (Yes/ NO)")
    if ans =='Yes' or ans =='yes':
        return True
    elif ans=='No' or ans =='no':      
        return False
```
